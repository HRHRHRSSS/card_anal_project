```python
import pandas as pd

def correlation_by_category_and_region(category, df_log, weather_df):
    correlations = {}
    for region in df_log['지역'].unique():
        # 주어진 분류와 지역에 해당하는 물품 필터링
        filtered_df = df_log[(df_log['분류'] == category) & (df_log['지역'] == region)]
        
        # '구매일시'를 기준으로 일자별로 행의 수를 계산
        daily_purchase_count = filtered_df.groupby(filtered_df['구매일시'].str[:10]).size().reset_index(name='구매건수')

        # 날짜 형식 변환
        daily_purchase_count['구매일시'] = pd.to_datetime(daily_purchase_count['구매일시'])

        # 새로운 데이터프레임을 기존의 날씨 데이터와 병합
        merged_df = pd.merge(daily_purchase_count, weather_df, left_on=['구매일시'], right_on=['일시'], how='inner')

        # 필요한 경우 날짜 및 시간 형식 변환
        merged_df['일시'] = pd.to_datetime(merged_df['일시'])

        # 상관 관계 계산
        correlation = merged_df[['구매건수', '평균기온(°C)', '일강수량(mm)', '평균 상대습도(%)', '평균 현지기압(hPa)', '일 최심신적설(cm)', '평균 전운량(1/10)']].corr()
        
        correlations[region] = correlation['구매건수']
    
    return correlations

# 예시: '분류'가 '과일'인 경우 모든 지역에 대한 상관 관계 계산
category_correlations_by_region = correlation_by_category_and_region('과일', df_log, weather_df)
for region, correlation in category_correlations_by_region.items():
    print(f"지역: {region}")
    print(correlation)
    print()
```
