```python
# 분류
# 분류명 입력형
# 연령대별 성별 분석 로직
category_input = input('분류명을 입력하시오(,로 구분) :')
category_list = category_input.split(',')

df_result = pd.DataFrame()

# 해당 품목에 대한 데이터 추출 from 로그데이터 (분자)
cond1 = (df1['분류'].isin(category_list))
temp = df1.loc[cond1]
for age in temp['ages'].unique().tolist() :
    temp_dict = {}
    cond2 = (temp['ages'] == age)
    cond3 = (temp['ma_fem_dv'] == '여성')
    target_num1 = len(temp.loc[cond2 & cond3]['cust'].unique()) # 20대 여성 
    cond4 = (temp['ma_fem_dv'] == '남성')
    target_num2 = len(temp.loc[cond2 & cond4]['cust'].unique()) # 20대 남성 
    
    # 고객데이터(분모)
    cond5 = (df_1['ma_fem_dv'] == '여성')
    cond6 = (df_1['ma_fem_dv'] == '남성')
    cond7 = (df_1['ages'] == age)
    total_num1 = len(df_1.loc[cond5&cond7])
    total_num2 = len(df_1.loc[cond6&cond7])
    
    women_ages_num = round(target_num1 / total_num1, 3)
    men_ages_num = round(target_num2 / total_num2, 3)
    
    temp_dict['여성'] = [women_ages_num]
    temp_dict['남성'] = [men_ages_num]
    dict1 = pd.DataFrame(temp_dict, index = [f'{age}'])
    df_result = pd.concat([df_result, dict1], axis = 0)

print('분류명 : ',category_input )
df_result
```
